```python
# Celda 5: leer los datos del repo con pandas
import pandas as pd

df = pd.read_csv("datos/ventas_ejemplo.csv")
df
```

```python
# Celda 6: graficar los datos del repo
tabla = df.pivot(index="mes", columns="region", values="ventas")
tabla = tabla.reindex(["Enero", "Febrero", "Marzo"])
tabla.plot(kind="bar", title="Ventas por región (datos del repo)")
plt.ylabel("Ventas")
plt.show()
```
